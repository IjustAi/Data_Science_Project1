```python
import matplotlib.pyplot as plt
import seaborn as sns

def visualize_relationship(data):
    plt.figure(figsize=(15,10))
    
    for i, column in enumerate(data.columns[:]):
        plt.subplot(4,5,i+1)
        sns.countplot(x=column, hue = 'num', data=data, palette='Set2')
        plt.title(f'{column} vs. Num')
        plt.xlabel(column)
        plt.ylabel('Count')
        
    plt.tight_layout()
    plt.show()
```

# Age

```python
# # Overlay KDE plots for 'age' comparing cases with and without heart disease
plt.figure(figsize=(12, 6))
sns.kdeplot(data[data['num'] > 0]['age'], color='skyblue', label='Heart Disease')
sns.kdeplot(data[data['num'] == 0]['age'], color='salmon', label='No Heart Disease')
plt.title('Kernel Density Estimate of Age for Patients with and without Heart Disease')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
# Boxplot of 'age' for cases with and without heart disease
plt.figure(figsize=(12, 8))
sns.boxplot(x='num', y='age', data=data, palette=['salmon', 'skyblue'])
plt.title('Boxplot of Age for Patients with and without Heart Disease')
plt.xlabel('Heart Disease (0: No, 1: Yes)')
plt.ylabel('Age')
plt.show()
```

```python
from scipy.stats import ttest_ind, mannwhitneyu

# Separate the data into groups where 'num' is 0 and 1.
group_0 = data[data['num'] == 0]['age']
group_1 = data[data['num'] == 1]['age']

# Independent sample t-test
t_stat, t_p_value = ttest_ind(group_0, group_1)

# Mann-Whitney U test (non-parametric test)
mwu_stat, mwu_p_value = mannwhitneyu(group_0, group_1)

# Display the results.
print(f"Independent t-test p-value: {t_p_value}")
print(f"Mann-Whitney U test p-value: {mwu_p_value}")
```

# SEX

```python
# Visualzie
plt.figure(figsize=(8, 6))
sns.countplot(x='sex', hue='num', data=data)
plt.title('Countplot of Binary Target Variable by Gender')
plt.xlabel('Sex (0 = Female, 1 = Male)')
plt.ylabel('Count')
plt.show()
```

```python
# Frequency of binary target variable by gender
female_counts = data[data['sex'] == 0]['num'].value_counts()
male_counts = data[data['sex'] == 1]['num'].value_counts()

# Visualize
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Female
axes[0].pie(female_counts, labels=female_counts.index, autopct='%1.1f%%', startangle=90, colors=['#ff9999', '#66b3ff'])
axes[0].set_title('Female: Distribution of Binary Target Variable')

# Male
axes[1].pie(male_counts, labels=male_counts.index, autopct='%1.1f%%', startangle=90, colors=['#ff9999', '#66b3ff'])
axes[1].set_title('Male: Distribution of Binary Target Variable')

plt.show()
```

```python
from scipy.stats import chi2_contingency

# Cross-tabulation between gender and binary target variables
cross_tab = pd.crosstab(data['sex'], data['num'])

# Perform a chi-square test
chi2, p_value, _, _ = chi2_contingency(cross_tab)

print(f"Chi-squared test statistic: {chi2}")
print(f"P-value: {p_value}")

# Significance test at significance level 0.05
alpha = 0.05
if p_value < alpha:
    print("Results: Reject the null hypothesis, there is a statistically significant difference between gender and binary target variables.")
else:
    print("Results: Null hypothesis accepted, there is no statistically significant difference between gender and binary target variables.")
```

# 최종결과

- 타겟변수와 상관관계 우선순위
    - ca > thal > oldpeak > thalack > cp
- slope와 oldpeak는 유의미한 상관관계를 가짐 / 하나의 변수로 합칠 가능성 있음
- 나이는 유의미하게 심장병에 영향을 주는 변수이며 60대가 심장병에 취약함
- 성별은 유의미하게 심장병에 영향을 주는 변수이며 남성이 더 심장병에 취약함
- Trestbps(안정시 혈압)은 유의미하게 심장병에 영향을 주는 변수이지만 그래프나 상관관계를 고려하였을 때 큰 설명력을 가지지는 못한다
- Chol(콜레스테롤)은 통계적으로 유의하지 않는다. 제거를 고려
- Fbs(공복혈당)은 통계적으로 유의하지 않는다. 제거를 고려
- Restecg(안정 심전도 결과)는 유의미하게 심장병에 영향을 주는 변수
- Exang(운동 유발 협심증)은 유의미하게 심장병에 영향을 주는 변수
- Oldpeak(운동에 의한 ST 감소)는 유의미하게 심장병에 영향을 주는 변수 
- Slope(운동 st 세그먼트의 기울기)는 유의미하게 심장병에 영향을 주는 변수
